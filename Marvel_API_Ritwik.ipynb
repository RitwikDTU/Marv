### Function to create dataframe containing Marvel Characters

```python
import requests
import json
import datetime as dt
import hashlib
from pprint import pprint as pp
import pandas as pd

def hash_generator(pub_key, priv_key, tstamp):

        public_key = pub_key
        private_key = priv_key
        timestamp = tstamp

        hash_md5 = hashlib.md5()
        hash_md5.update(f'{timestamp}{private_key}{public_key}'.encode('utf-8'))
        hashed_params = hash_md5.hexdigest()
        
        return hashed_params


def get_dataframe(pub_key, priv_key, tstamp = dt.datetime.now().strftime('%Y-%m-%d%H:%M:%S') ):

        character_data_list =[]

        for i in range(16):
            params = {'ts':tstamp, 'apikey':pub_key, 'hash':hash_generator(pub_key,priv_key,tstamp),'limit':100, 'offset':100*i}
            response = requests.get('https://gateway.marvel.com:443/v1/public/characters',params)

            result = response.json()

            # Taking a copy of the result json file
            result_copy = result

            # Extracting characters' information from the json object and storing them into a list 

            for j in range(len(result_copy['data']['results'])):
                l = []
                l.append(result_copy['data']['results'][j]['id'])
                l.append(result_copy['data']['results'][j]['name'])
                l.append(result_copy['data']['results'][j]['events']['available'])
                l.append(result_copy['data']['results'][j]['series']['available'])
                l.append(result_copy['data']['results'][j]['comics']['available'])
                l.append(result_copy['data']['results'][j]['stories']['available'])

                character_data_list.append(l) # List containing all characters' information
                
            print('Fetched ',100*(i+1), 'characters')
        

        # Converting list to final dataframe
        character_dataframe = pd.DataFrame.from_records(character_data_list)
        character_dataframe.rename(columns = {0:'Character Id',1:'Name',2:'Event Appearances',3:'Series Appearances',4:'Comics Appearances',5:'Stories Appearances'}, inplace = True)
        character_dataframe.set_index('Character Id',inplace=True)


        return character_dataframe
```

```python
public_key = '00a4a193348e48ae938803b520bdeac5'
private_key = '231229fd3e703b82b090bd22ae05263f191c9e03'
df = get_dataframe(public_key,private_key)
```

### Function for filtering data based on different conditions

```python
'''

Available filters :

    A. For numerical columns:
    
        1. greater than n : find records where column value is greater than specified number n
        2. less than n : find records where column value is lesser than specified number n
        3. equal to n : find records where column value is equal than specified number n
        4. min : find records where column value is equal to minimum value of that column
        5. max : find records where column value is equal to maximum value of that column
        
    B. For String columns:
    
        1. startsWith string : find records where name starts with specified string
        2. searchStr string : find records where name contains specified string 
    
''' 

def filter_column(character_df, column_name, filter_cond):
    
    df = character_df
    col = column_name
    fltr = filter_cond
    
    t_df = pd.DataFrame()
    
    if 'min' in fltr:
        t_df = df.loc[df[col] == min(df[col])]
        return t_df
    
    if 'max' in fltr:
        t_df = df.loc[df[col] == max(df[col])]
        return t_df  
        
    if 'greater than' in fltr:
        fltr = fltr.strip()
        n = [int(i) for i in fltr.split() if i.isdigit()][0]
        t_df = df.loc[df[col]>n]
        
        return t_df
    
    if 'less than' in fltr:
        fltr = fltr.strip()
        n = [int(i) for i in fltr.split() if i.isdigit()][0]
        t_df = df.loc[df[col]<n]
        
        return t_df
    
    if 'equal to' in fltr:
        fltr = fltr.strip()
        n = [int(i) for i in fltr.split() if i.isdigit()][0]
        t_df = df.loc[df[col]==n]
        
        return t_df
    
    if 'startsWith' in fltr:
        fltr = fltr.strip()
        char_index = fltr.find('startsWith')+len('startsWith')+1
        string = fltr[char_index:]
        string = string.strip()
        t_df = df.copy()
        t_df['new_col'] = 0
        for i in range(len(df)):
            if df[col].iloc[i].startswith(string) == True:
                t_df.iloc[i,-1] = 1
                
        t_df = t_df.loc[t_df['new_col'] == 1]
        
        return t_df.iloc[:,:-1]
    
    if 'searchStr' in fltr:
        fltr = fltr.strip()
        char_index = fltr.find('searchStr')+len('searchStr')+1
        string = fltr[char_index:]
        string = string.strip()
        t_df = df.copy()
        t_df['new_col'] = 0
        for i in range(len(df)):
            if df[col].iloc[i].find(string) != -1:
                t_df.iloc[i,-1] = 1
                
        t_df = t_df.loc[t_df['new_col'] == 1]
        
        return t_df.iloc[:,:-1]
    
```
